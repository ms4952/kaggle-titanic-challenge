```python
import pandas as pd
import numpy as np
data=pd.read_csv('./titanic/train.csv')
df=pd.DataFrame(data,columns=['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked'])
df.head()
```

```python
df.Cabin=np.where(df.Cabin.isnull(),'N',df['Cabin'])
#df['Cabin']=df['Cabin'].astype('|S')
df.Sex=df.Sex.astype('|S')
df.Embarked=np.where(df.Embarked.isnull(),'N',df['Embarked'])
df.Age=np.where(df.Age.isnull(),df['Age'].mean(),df['Age'])
#df.Age=df.Age.astype(int)
for i in range(0,len(df)):
    df['Cabin'][i]=df['Cabin'][i][0]
df.head()
```

```python
df=pd.get_dummies(df, columns=["Sex","Cabin","Embarked"])

df.head()
```

```python
df.to_csv('./titanic/wrangling.csv',index = False)
```

```python
from sklearn.model_selection import train_test_split
#X = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
X=df.drop(['Survived'], axis = 1)
Y=df.Survived
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=.3,random_state=0)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
clf = AdaBoostClassifier(n_estimators=50,learning_rate=1)
clf.fit(X_train,y_train)

gb=GradientBoostingClassifier(n_estimators=100, learning_rate = 1, max_features=3, max_depth = 2, random_state = 0)
gb.fit(X_train,y_train)

y_pred = gb.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```
