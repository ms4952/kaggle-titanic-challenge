```python
import pandas as pd
import numpy as np
data=pd.read_csv('./titanic/train.csv')
data1=pd.read_csv('./titanic/test.csv')
df=pd.DataFrame(data,columns=['Survived','Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked'])
df_test=pd.DataFrame(data1,columns=['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked'])
df_test.head()
```

```python
df.Cabin=np.where(df.Cabin.isnull(),'N',df['Cabin'])
#df['Cabin']=df['Cabin'].astype('|S')
df.Sex=df.Sex.astype('|S')
df.Embarked=np.where(df.Embarked.isnull(),'N',df['Embarked'])
df.Age=np.where(df.Age.isnull(),df['Age'].mean(),df['Age'])
df.Age=df.Age.astype(int)
for i in range(0,len(df)):
    df['Cabin'][i]=df['Cabin'][i][0]
df.head()
```

```python
df_test.Cabin=np.where(df_test.Cabin.isnull(),'N',df_test['Cabin'])
#df_test['Cabin']=df_test['Cabin'].astype('|S')
df_test.Sex=df_test.Sex.astype('|S')
df_test.Embarked=np.where(df_test.Embarked.isnull(),'N',df_test['Embarked'])
df_test.Age=np.where(df_test.Age.isnull(),df_test['Age'].mean(),df_test['Age'])
df_test.Fare=np.where(df_test.Fare.isnull(),df_test['Fare'].mean(),df_test['Fare'])
df_test.Age=df.Age.astype(int)
for i in range(0,len(df_test)):
    df_test['Cabin'][i]=df_test['Cabin'][i][0]
df_test.isnull().sum()
```

```python
df=pd.get_dummies(df, columns=["Sex","Cabin","Embarked"])

df.head()
```

```python
df_test=pd.get_dummies(df_test, columns=["Sex","Cabin","Embarked"])
df_test["Cabin_T"]=0
df_test["Embarked_N"]=0
df_test.head()
```

```python
from sklearn.model_selection import train_test_split
#X = df[['Pclass','Sex','Age','SibSp','Parch','Fare','Cabin','Embarked']]
X=df.drop(['Survived'], axis = 1)
Y=df.Survived
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=.3,random_state=1)
print(df_test.head())
print(X.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
# Create the model with 100 trees
clf = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
clf.fit(X,Y)
gb=GradientBoostingClassifier(n_estimators=20, learning_rate = .75, max_features=2, max_depth = 2, random_state = 0)
gb.fit(X_train,y_train)

y_pred = gb.predict(df_test)
```

```python
len(y_pred)
data3=pd.read_csv('./titanic/test.csv')
df_submit=pd.DataFrame(data3,columns=["PassengerId"])
df_submit["Survived"]=y_pred
df_submit.to_csv('./titanic/submission.csv',index = False)
```
